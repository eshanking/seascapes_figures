```python
from fears.population import Population
import matplotlib.pyplot as plt
import numpy as np
from fears.utils import plotter,fitness

p = Population(fitness_data='from_file')
```

```python
xdata = np.logspace(-4,5,100)
# append 0 to the left
xdata = np.append(0,xdata)

cc = plotter.gen_color_cycler()

fig,ax = plt.subplots(figsize=(5,4))

ax.set_prop_cycle(cc)

for i in range(16):
    ydata = fitness.gen_fitness(p,i,xdata)
    ax.plot(xdata,ydata,label='{}'.format(p.int_to_binary(i)),linewidth=2)

ax.set_xscale('symlog',linthresh=1e-4)   

ax.set_xlabel('Drug Concentration (g/ml)',fontsize=14)
ax.set_ylabel('Growth Rate (hr$^{-1}$)',fontsize=14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend(ncol=1,frameon=False,fontsize=10,loc=(1.05,-0.1))

fig.savefig('../figures/seascape_fig.svg',bbox_inches='tight')
```
